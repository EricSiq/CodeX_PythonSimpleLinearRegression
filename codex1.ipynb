```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df=pd.read_csv('/kaggle/input/salary-dataset-simple-linear-regression/Salary_dataset.csv')
```

```python
#step 2
plt.scatter(df['YearsExperience'] , df['Salary'])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

```python
#Step 3
X=df[['YearsExperience']]
Y=df['Salary']
```

```python
#Step 4
model = LinearRegression()
model.fit(X,Y) #calculates best fitting relation between x and y
slope = model.coef_[0] #represents coeff of x in y=mx+b
intercept=model.intercept_ #point where regression line crosses y axis
```

```python
#Step 5
def linear_regression(x):
    return slope * x + intercept
```

```python
#Step 6
plt.scatter(df['YearsExperience'],df['Salary'],label='Data Points')
# Assuming df is your DataFrame containing the data
plt.scatter(df['YearsExperience'], df['Salary'], label='Data Points', color='blue')
plt.plot(df['YearsExperience'], [linear_regression(x1) for x1 in df['YearsExperience']], color='red', label='Regression Line')
plt.xlabel('X:Years of Experience')
plt.ylabel('Y:Salary')
plt.title("Salary vs. Years of Experience")
plt.legend()
plt.show()

```

```python

new_x = 6
prediction = linear_regression(new_x)
print(f"Prediction for x = {new_x}: {prediction}")

new_x_df = pd.DataFrame({'YearsExperience': [new_x]})
predicted_value = model.predict(new_x_df)
print(predicted_value)
```
